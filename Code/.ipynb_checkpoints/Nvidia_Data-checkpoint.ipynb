```python
# Change the Date format

df['Date']= pd.to_datetime(df['Date'],utc=True).dt.date

df.head()

```

```python
# Dataframe with rows when dividends are paid

dividends = df[df['Dividends'] != 0]
print(dividends)
```

```python
# Dataframe with rows when dividends are paid

stock_split = df[df['Stock Splits'] != 0]
print(stock_split)
```

```python
# Drop the Dividends and Stock Splits columns

df.drop(columns=['Dividends','Stock Splits'], inplace = True)

df.head()
```

```python
# Calculate Simple Moving Average

df['SMA'] = df['Close'].rolling(window = 30).mean()
df.fillna(0, inplace = True)
df.head()
```

```python
# Export dataframes to csv file

df.to_csv(r"C:\Users\ragul\Downloads\Nvidia Project\Datasets\nvidia_stock_preprocessed.csv")
dividends.to_csv(r"C:\Users\ragul\Downloads\Nvidia Project\Datasets\nvidia_dividends.csv")
stock_split.to_csv(r"C:\Users\ragul\Downloads\Nvidia Project\Datasets\nvidia_stock_splits.csv")
```

```python
# Calculate % Yearly return 

# Create a Year column

df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Group by year

yearly_prices = df.groupby('Year').agg(
    Start_Price=('Close', 'first'),
    End_Price=('Close', 'last')
).reset_index()

# Calculate yearly % return

yearly_prices['%Return'] = ((yearly_prices['End_Price'] - yearly_prices['Start_Price']) / yearly_prices['Start_Price']) * 100 

yearly_prices['Positive_Return'] = yearly_prices['%Return'].apply(lambda x: x if x > 0 else 0)
yearly_prices['Negative_Return'] = yearly_prices['%Return'].apply(lambda x: x if x < 0 else 0)

print(yearly_prices)
# Save to CSV 
yearly_prices.to_csv(r"C:\Users\ragul\Downloads\Nvidia Project\Datasets\nvidia_yearly_returns.csv", index=False)
```

```python
# Calculate Annual Percentage return

# Create year column

nasdaq['Year'] = nasdaq['Date'].dt.year

# Group by year

yearly_prices = nasdaq.groupby('Year').agg(
    Start_Price=('Close', 'first'),
    End_Price=('Close', 'last')
).reset_index()

# Calculate yearly returns 

yearly_prices['Yearly_Return_%'] = ((yearly_prices['End_Price'] - yearly_prices['Start_Price']) / yearly_prices['Start_Price']) * 100
yearly_prices['Positive_Return'] = yearly_prices['Yearly_Return_%'].apply(lambda x: x if x > 0 else 0)
yearly_prices['Negative_Return'] = yearly_prices['Yearly_Return_%'].apply(lambda x: x if x < 0 else 0)

yearly_prices
```

```python
# Create a dataframe with last 3 years data

df = pd.read_csv(r"C:\Users\ragul\Downloads\Nvidia Project\Datasets\nvidia_stock_data.csv")

df_last_three = df.tail(806)

df_last_three.head()
```

```python
# Drop unwanted columns 

df_last_three = df_last_three.drop(columns = ['Dividends', 'Stock Splits', 'Low','High','Open'])

df_last_three
```

```python
# Change date format

df_last_three['Date'] = pd.to_datetime(df_last_three['Date'],utc=True).dt.date
df_last_three
```

```python
# Calculate Moving Averages

df_last_three['MA20'] = df_last_three['Close'].rolling(window=20).mean()

df_last_three['MA50'] = df_last_three['Close'].rolling(window=50).mean()

df_last_three.head()
```

```python
# Drop first 50 rows 

df_last_three = df_last_three.tail(756)

df_last_three
```

```python
# Create new columns to add signal and cross 

df_last_three = df_last_three.copy()

df_last_three['Signal'] = 0
df_last_three.loc[df_last_three['MA20'] > df_last_three['MA50'], 'Signal'] = 1
df_last_three['Cross'] = df_last_three['Signal'].diff()

df_last_three.head()
```

```python
# Create a column for marker

df_last_three['GoldenCross'] = np.where(df_last_three['Cross'] == 1, df_last_three['MA20'], np.nan)
df_last_three['DeathCross']  = np.where(df_last_three['Cross'] == -1, df_last_three['MA20'], np.nan)

df_last_three.head()
```

```python
# Save to csv

df_last_three.to_csv(r'C:\Users\ragul\Downloads\Nvidia Project\Datasets\nvidia_moving_averages.csv')
```
